```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
from scipy.signal import find_peaks_cwt
import pandas as pd
import os
import datetime
import itertools
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv('result_CA4000431_000000681000011800.csv')
df.columns = [c.lower() for c in df.columns]
df['fscl_wk_end_dt'] = pd.to_datetime(df['fscl_wk_end_dt'])
df.set_index('fscl_wk_end_dt', inplace = True)
orig_df = df.copy()
df = df[['ty_div_sales_units']]
df
```

```python
df.plot(figsize = (20,10))
```

```python
x = df['ty_div_sales_units'].values
x_neg = x * -1

peaks_pos_index, _ = find_peaks(x)
figure(figsize=(20, 10), dpi=80)
plt.plot(x)

plt.plot(peaks_pos_index, x[peaks_pos_index], "x", color='green')
plt.plot(np.zeros_like(df['ty_div_sales_units'].values), "--", color="gray")
peaks_neg_index, _ = find_peaks(x_neg)
plt.plot(peaks_neg_index, x[peaks_neg_index], "*", color='red')
plt.show()
```
